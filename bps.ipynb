```python
import pandas as pd
import statsmodels.formula.api as smf

# 1. Load Data
df = agg[agg['country'] == 'United States']

# 2. Determine the Schedule (Treatment vs Control Assignment)
# We look at the Test Period (test_period=1) to see which hours are assigned to Ads ON vs OFF.
test_data = df[df['test_period'] == 1]

# Check if each hour has a consistent 'ads_on' status
# We group by hour and calculate the mean of 'ads_on'. 
# If mean is 0, it's always OFF. If mean is 1, it's always ON. If between, it rotates.
schedule = test_data.groupby('hour')['ads_on'].mean()

print("Schedule Consistency Check (Mean Ads_On per Hour):")
print(schedule)

# 3. Assign Groups based on the Schedule
# If the schedule is consistent (values are either 0.0 or 1.0), we can map this back to the pre-period.
# We create a 'treatment_group' flag. 
# Let's define: 
# Group = 1 (Treatment / Changed): Hours where Ads turned OFF (ads_on = 0 in test)
# Group = 0 (Control / Unchanged): Hours where Ads stayed ON (ads_on = 1 in test)

# Create a mapping dictionary from the schedule
# Note: If schedule is mixed (e.g. 0.5), DiD is harder. We assume fixed here.
hour_to_group = (schedule == 0).astype(int).to_dict() 
# result: 1 if ads_on is 0 (Changed/OFF), 0 if ads_on is 1 (Unchanged/ON)

df['is_ads_off_group'] = df['hour'].map(hour_to_group)

# 4. Filter Data for Validity
# Ensure we only keep hours that were strictly ON or strictly OFF in the schedule
valid_hours = schedule[schedule.isin([0.0, 1.0])].index
df_did = df[df['hour'].isin(valid_hours)].copy()

# 5. Run Difference-in-Differences Regression
# Formula: trials ~ test_period + is_ads_off_group + test_period:is_ads_off_group
# Interpretation:
# - test_period: Baseline change in traffic over time (for the Unchanged group)
# - is_ads_off_group: Baseline difference between the two sets of hours
# - test_period:is_ads_off_group (Interaction): The DiD Estimator. The effect of turning ads OFF.

model = smf.ols(
    formula="trials ~ test_period * is_ads_off_group + C(country) + C(day_of_week) ",
    data=df_did
).fit()

print("\n--- Difference-in-Differences Results ---")
print(model.summary())
```
